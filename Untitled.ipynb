```python
total = df[df['location'] == 'United States']
```

```python
total.head()
```

```python
total.describe().style.set_table_styles(
[{'selector': 'tr:hover',
  'props': [('background-color', 'yellow')]}]
)
```

```python
total.loc[:,'total_date'] = total.groupby('date')['total_cases'].transform('sum') 
```

```python
plot_total= total.drop_duplicates('date', keep = 'first', inplace = False) # dropping the duplicates but keeping just the first 
```

```python
plt.figure(figsize=(14,10))
#plot_total_rape.plot(legend=False)
x = plot_total['date']
y = plot_total['total_date']
plt.plot(x,y)
plt.title('Number of cases over the year 2020 in USA ',color = 'blue',fontsize=14)
plt.xlabel('Date')
plt.ylabel('Total number of cases')
plt.grid(True)
plt.show()
```
